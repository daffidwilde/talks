```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy.stats import linregress

%matplotlib inline
```

```python
df = pd.read_csv("anscombe.csv")

fig, (top, bottom) = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), dpi=300)

for i, (group, data) in enumerate(df.groupby("dataset")):
    axes = top if i in [0, 1] else bottom
    ax = axes[i % 2]

    xs = np.linspace(-1, 21, 100)
    slope, intercept, *_ = linregress(data["x"], data["y"])
    ax.plot(xs, slope * xs + intercept, zorder=-1)
    
    ax.scatter(data["x"], data["y"], color="darkorange")
    ax.set_title(group, fontsize=16)
    ax.set_xlim(3, 20)
    ax.set_ylim(2, 14)
    ax.set_aspect("equal")

plt.tight_layout(pad=2)
plt.savefig("../img/anscombes.pdf", transparent=True)
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(12, 8), dpi=300)

for i, file in enumerate(["simpsons.csv", "datasaurus.csv", "random_cloud.csv"]):

    data = pd.read_csv(file)
    ax = axes[i]
    
    ax.scatter(data["x"], data["y"], marker=".")
    ax.set_xlim(-10, 110)
    ax.set_ylim(-10, 110)
    ax.set_aspect("equal")

plt.tight_layout(pad=2)
plt.savefig("../img/datasaurus.pdf", transparent=True)
```
